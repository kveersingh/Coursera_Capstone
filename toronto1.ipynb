```python
import pandas as pd
```

creating data frame

```python
table=pd.DataFrame(columns=['postalcode','borough','neighborhood'])
```

creating empty list

```python
postalcode=list()
borough=list()
neighborhood=list()
```

getting data in the lists created above

```python
s=soup.select('tr')

for i in s:
    i=i.text
    i=i.split('\n')
    if len(i)==5:
        i=i[1:4]
        print(i)
        postalcode.append(i[0])
        borough.append(i[1])
        neighborhood.append(i[2])
        
    else:
        pass                
```

to remove first value from every list

```python
del postalcode[0]
del neighborhood[0]
del borough[0]
```

geting data in data frame from list

```python
table['borough']=borough
table['neighborhood']=neighborhood
table['postalcode']=postalcode
```

```python
table= table[table['borough']!='Not assigned'].reset_index() #removeing rows in which borough is not assigned
```

joining rows with common postal code and borough

```python
nt= table.groupby(by=['postalcode','borough'])['neighborhood'].apply(lambda x: " , ".join(x.astype(str))).reset_index()
```
